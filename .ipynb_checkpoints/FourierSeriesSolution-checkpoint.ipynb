```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 20
signal = np.zeros(N)
signal[:10] = 1
plt.plot(signal)
plt.show()
```

```python
transform = np.fft.fft(signal)
plt.plot(np.abs(transform))
plt.show()
```

```python
powers = 2j*np.pi*np.linspace(0, N, 1000)/N # approximate a continous signal

reconstructed = np.zeros_like(powers)
for k in range(10):
    reconstructed += transform[k]*np.exp(k*powers)/N
    plt.plot(np.real(reconstructed), label=f"k={k+1}")
plt.legend()
plt.show()
```
