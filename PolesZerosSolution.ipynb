```python
import control
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create transfer function
H = control.TransferFunction(np.array([1, -4*np.cos(np.pi/4), 4]), np.array([1, -0.5*np.cos(3*np.pi/4), 1/16]))
H
```

```python
# plot poles and zeros
control.pzmap(H) 

# Add circle
theta = np.linspace(0, 2*np.pi, 100)
x1 = np.cos(theta)
x2 = np.sin(theta)
plt.plot(x1, x2, linestyle=":")
plt.show()
```

```python
zero = np.exp(1j*np.pi/4)
pole = np.exp(3j*np.pi/4)
Ha = control.TransferFunction(np.array([1, -2*zero]), np.array([1, -0.5*zero]))*control.TransferFunction(np.array([1, -2*np.conj(zero)]), np.array([1, -0.5*np.conj(zero)]))
Hm = control.TransferFunction(np.array([1, -0.5*zero]), np.array([1, -0.25*pole]))*control.TransferFunction(np.array([1, -0.5*np.conj(zero)]), np.array([1, -0.25*np.conj(pole)]))
```

```python
# plot poles and zeros
control.pzmap(Ha) 

# Add circle
theta = np.linspace(0, 2*np.pi, 100)
x1 = np.cos(theta)
x2 = np.sin(theta)
plt.plot(x1, x2, linestyle=":")

# plot poles and zeros
control.pzmap(Hm) 

# Add circle
theta = np.linspace(0, 2*np.pi, 100)
x1 = np.cos(theta)
x2 = np.sin(theta)
plt.plot(x1, x2, linestyle=":")
plt.show()
```
