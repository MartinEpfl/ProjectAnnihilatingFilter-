```python
P = np.array([[0.8, 0.2, 0.1],
              [0.1, 0.6, 0.2],
              [0.1, 0.2, 0.7]])
```

```python
state=np.array([[1.0, 0.0, 0.0]])
states = np.zeros((100,3))
states[0] = state
print(P)
for x in range(100):
  states[x]= np.random.multinomial(1, states[x-1]@P)
  print(states[x])
  

```
